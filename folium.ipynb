```python
import folium
import pandas as pd 
```

```python
df = pd.DataFrame(pd.read_csv('8,9,2023.csv'))
df.dtypes
df
```

```python
n = folium.Map(location=(df['Latitude'][0],df['Longtitude'][0]),tiles="OpenStreetMap", zoom_start=25)

for i in range(0,len(df)) :
    folium.Marker(
        location=[df.iloc[i]['Latitude'], df.iloc[i]['Longtitude']],
        popup=f"""id = {df.iloc[i]['id']}\nCoordinaate {df.iloc[i]['Latitude']},{df.iloc[i]['Longtitude']}\,
        time = {df.iloc[i]['Time']}
        
        """,draggable=False
    ).add_to(n)

n
    
```
